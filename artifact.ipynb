# Cajal: Artifact Overview

**Cajal** is a differentiable compiler that translates programs written in a simply-typed lambda calculus into PyTorch tensors (multilinear maps). This encoding preserves type structure and supports automatic differentiation, enabling gradient-based learning of program semantics.

This notebook demonstrates the core pipeline:
1. Constructing terms using `cajal.syntax`
2. Type-checking with `cajal.typing`
3. Evaluating with `cajal.evaluating`
4. Compiling to tensors with `cajal.compiling`

```python
from cajal.syntax import TmFun, TmIf, TmVar, TmTrue, TmFalse, TyBool

# λx:Bool. if x then false else true  (boolean NOT)
not_term = TmFun("x", TyBool(), TmIf(TmVar("x"), TmFalse(), TmTrue()))
print(not_term)
```
